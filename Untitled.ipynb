## What is the diffraction like in near field (continue with Sandra's picture in the last lecture)?

* A simulation of slits diffraction
    
    * Slits are on the left side wall and the wave is propagating from left to right.
    
    * wavelength = $\lambda$
    
    * slit distance = 2$\lambda$
    
    * picture width = 30$\lambda$

```python
import matplotlib.pyplot as plt
from matplotlib.axes import Axes
from matplotlib.gridspec import GridSpec
import numpy as np

plt.style.use("group-mpl.txt")
```

```python
class SlitsDiffrac:
    
    def __init__(self, wavelength: float, x_grid: np.array, y_grid: np.array, x_slits: list, y_slits: list):
        self._wavelength = wavelength
        self._x_grid = x_grid
        self._y_grid = y_grid
        self._xx, self._yy = np.meshgrid(x_grid, y_grid)
        self._x_slits = x_slits
        self._y_slits = y_slits
        self._e_field = self._calc_e_field()
        self._i_field = self._e_field * self._e_field
        
    def _calc_e_field(self) -> np.array:
        e = np.zeros_like(self._xx)
        for x, y in zip(self._x_slits, self._y_slits):
            r = np.sqrt((self._xx - x) ** 2 + (self._yy - y) ** 2)
            e += np.divide(np.cos(2 * np.pi * r / self._wavelength), r, out=np.zeros_like(r), where=r!=0)
        return e
    
    def add_slit(self, x: float, y: float):
        self._x_slits.append(x)
        self._y_slits.append(y)
        self._e_field = self._calc_e_field()
        self._i_field = self._e_field * self._e_field
    
    def vis_i_field(self, ax: Axes, **kwargs):
        img = ax.imshow(self._i_field, **kwargs)
        xticks = ax.get_xticks()
        ax.set_title("{} slits".format(len(self._x_slits)))
        ax.set_xticklabels([])
        ax.set_xlabel("x")
        yticks = ax.get_yticks()
        ax.set_yticklabels([])
        ax.set_ylabel("y")
```

```python
d = 1.5
sd = SlitsDiffrac(
    wavelength=0.05,
    x_grid=np.linspace(0, d, 512),
    y_grid=np.linspace(0.5 - d / 2., 0.5 + d / 2., 512),
    x_slits=[],
    y_slits=[]
)
fig = plt.figure(figsize=(5 * 3, 5 * 2))
grids = iter(GridSpec(2, 3))
for x, y in [(0., 0.5), (0., 0.6), (0., 0.4), (0., 0.7), (0., 0.3), (0., 0.8)]:
    sd.add_slit(x, y)
    ax = fig.add_subplot(next(grids))
    sd.vis_i_field(ax, vmin=0, vmax=60)
del fig, ax, grids, sd, d
```

## Why can we use Fourier transformation to get PDF?

$$
E_{out} = \sum_i e^{- i \boldsymbol{k} \cdot (\boldsymbol{r} - \boldsymbol{r_i})}\\
= e^{- i \boldsymbol{k} \cdot \boldsymbol{r}} \sum_i e^{i \boldsymbol{k} \cdot \boldsymbol{r_i}}\\
= e^{- i \boldsymbol{k} \cdot \boldsymbol{r}} \mathscr{F} (\sum_i \delta(\boldsymbol{r} - \boldsymbol{r_i})) \\
= e^{- i \boldsymbol{k} \cdot \boldsymbol{r}} \mathscr{F}(T(\boldsymbol{r}))
$$

output electric field = $\mathscr{F}$(the window function) * plane wave field

In x-ray diffraction, the window function is atom density $T(\boldsymbol{r}) = \rho(\boldsymbol{r})$.

## Why can we use Fourier transformation to get PDF?

$$
I_{out} = E_{out} E^*_{out}\\
= \mathscr{F}(\rho(\boldsymbol{r})) \mathscr{F}^{*}(\rho(\boldsymbol{r}))\\
= \mathscr{F}(\rho(\boldsymbol{r})) \mathscr{F}(\rho(-\boldsymbol{r}))\\
= \mathscr{F}(\rho(\boldsymbol{r}) * \rho(-\boldsymbol{r}))\\
= \mathscr{F}(P(\boldsymbol{u}))
$$

output light intensity field = $\mathscr{F}$(the Patterson function)

## Why can we use Fourier transformation to get PDF?

$$
P(\boldsymbol{u}) = \rho(\boldsymbol{r}) * \rho(-\boldsymbol{r})\\
= \sum_i \delta(\boldsymbol{r} - \boldsymbol{R_i}) * \sum_j \delta(- \boldsymbol{r} - \boldsymbol{R_j})\\
= \sum_i \sum_j \delta(\boldsymbol{r} - (\boldsymbol{R_i} - \boldsymbol{R_j}))\\
= \sum_i \sum_j \delta(\boldsymbol{r} - \boldsymbol{r}_{ij})
$$

Patterson function = the distribution function of atomic pair distances

## The definition of R(r)

Radial distribution function: how many pairs of atoms are there at distance $r$ in a unit volume.

$$\large R(r) = \frac{1}{N} \sum_{u} \sum_{v} \frac{f_u f_v}{\bar{f}^2} \delta(r - r_{uv})$$

The integral is the number of atoms in the structure.

$$\large \int_{0}^{\infty} R(r) dr = N$$

## The definition of G(r)

Pair distribution function: how different is the density of atomic pairs at distance $r$ from the average.

$$\large G(r) = \frac{R(r)}{r} - 4{\pi}\rho_0r \text{, in 3D space}$$

$$\large G(r) = \frac{R(r)}{r} - 2{\pi}\rho_0 \text{, in 2D space}$$

$$\large G(r) = \frac{R(r)}{r} - \frac{2\rho_0}{r} \text{, in 1D space}$$

The integral is zero.

$$\large \int_{0}^{\infty} G(r) r dr = 0$$

```python
def Gaussian(x: np.array, a: float, mu: float, sigma: float):
    return a / (sigma * np.sqrt(2 * np.pi)) * np.exp(- 0.5 * ((x - mu) / sigma) ** 2)
    

class PairDistances:
    
    def __init__(self, distances: np.array, counts: np.array, std: float = 1., rho0: float = 1.):
        self._distances = distances
        self._counts = counts
        self._std = std
        self._rho0 = rho0
       
    @property
    def rho0(self):
        return self._rho0
    
    @rho0.setter
    def rho0(self, value: float):
        self._rho0 = value
    
    @property
    def std(self):
        return self._std
    
    @std.setter
    def std(self, value: float):
        self._std = value
        
    @classmethod
    def calculate(cls, x: np.array, y: np.array, precision: int = 3):
        r = np.round(np.sqrt(x ** 2 + y ** 2), precision)
        r = r[r > 0.]  # remove zero
        uniques, counts = np.unique(r, return_counts=True)
        return cls(uniques, counts)
    
    @property
    def distances(self):
        return self._distances
    
    @property
    def counts(self):
        return self._counts
    
    def yield_pairs(self, n: int):
        for i, (d, c) in enumerate(zip(self._distances, self._counts)):
            if i >= n:
                break
            yield d, c
            
    def yield_gaussian(self, x_grid: np.array, n: int):
        for d, c in self.yield_pairs(n):
            yield Gaussian(x_grid, c, d, self._std)
            
    def radial_distribution_function(self, x_grid: np.array, n: int) -> np.array:
        return sum(self.yield_gaussian(x_grid, n))
        
    def pair_distribution_function(self, x_grid: np.array, n: int) -> np.array:
        f0 = self.radial_distribution_function(x_grid, n)
        return np.divide(f0, x_grid, out=np.zeros_like(x_grid), where=x_grid!=0) - 2 * np.pi * self._rho0
    
    def linear_baseline(self, x_grid: np.array) -> np.array:
        return 2 * np.pi * self._rho0 * x_grid


def color(i: int) -> str:
    cycle = ['red', 'green', 'blue']
    return cycle[i % len(cycle)]


def fftnoise(f: np.array, freq: float) -> np.array:
    f = np.array(f, dtype='complex')
    Np = (len(f) - 1) // 2
    phases = np.random.rand(Np) * 2 * np.pi * freq
    phases = np.cos(phases) + 1j * np.sin(phases)
    f[1:Np+1] *= phases
    f[-1:-1-Np:-1] = np.conj(f[1:Np+1])
    return np.fft.ifft(f).real


class SquareCrystal:
    
    def __init__(self, n: int, a: float, u: float):
        x = np.arange(-n, n + 1) * a
        y = np.arange(-n, n + 1) * a
        xx, yy = np.meshgrid(x, y)
        self._x = xx.flatten()
        self._y = yy.flatten()
        self._n = n
        self._a = a
        self._u = u
        self._pd = PairDistances.calculate(xx, yy)
        self._pd.std = u
        self._pd.rho0 = 1 / a ** 2
        
    @property
    def pd(self):
        return self._pd
        
    def visualize(self, ax: Axes, **kwargs):
        ax.scatter(self._x, self._y, **kwargs)
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        
    def view_unit_cell(self, ax: Axes, **kwargs):
        a = self._a
        x = [0., a, 0., a]
        y = [0., 0., a, a]
        c = ['r', 'b', 'b', 'b']
        ax.scatter(x, y, c=c, **kwargs)
        ax.vlines([0., a], ymin=[0., 0.], ymax=[a, a], ls='--', color='black')
        ax.hlines([0., a], xmin=[0., 0.], xmax=[a, a], ls='--', color='black')
        ax.set_xlabel("x")
        ax.set_ylabel("y")
    
    def circles(self, ax: Axes, n: int, **kwargs):
        pd = self._pd
        for i, (d, _) in enumerate(pd.yield_pairs(n)):
            circle = plt.Circle((0, 0), d, fill=False, color=color(i), **kwargs)
            ax.add_patch(circle)
        return ax
    
    def radial_distribution_delta(self, ax: Axes, n: int):
        colors = [color(i) for i in range(len(self._pd.counts))]
        ax.vlines(self._pd.distances[:n], ymin=0, ymax=self._pd.counts[:n], colors=colors[:n])
        ax.set_xlabel("r")
        ax.set_ylabel("Count")
        ax.set_ylim(0, np.max(self._pd.counts) + 1)
        
    def _xgrid(self, n_peak: int, n_grid: int) -> np.array:
        pd = self._pd
        dmax = np.max(pd.distances[:n_peak])
        dmin = np.min(pd.distances[:n_peak])
        x_grid = np.linspace(dmin - 3 * pd.std, dmax + 3 * pd.std, n_grid)
        return x_grid
    
    def _xgrid_from_zero(self, n_peak: int, n_grid: int) -> np.array:
        pd = self._pd
        dmax = np.max(pd.distances[:n_peak])
        x_grid = np.linspace(0, dmax + 3 * pd.std, n_grid)
        return x_grid
        
    def radial_distribution_gaussian(self, ax: Axes, n_peak: int, n_grid: int = 501):
        x_grid = self._xgrid(n_peak, n_grid)
        y_sum = np.zeros_like(x_grid)
        for i, y in enumerate(self._pd.yield_gaussian(x_grid, n_peak)):
            y_sum += y
            ax.plot(x_grid, y, color=color(i))
            ax.fill_between(x_grid, y, color=color(i), alpha=0.3)
        ax.plot(x_grid, y_sum, '--', c='black')
        ax.set_ylim(0, None)
        ax.set_xlabel("r")
        ax.set_ylabel("R")
        
    def pair_distribution_function(self, ax: Axes, n_peak: int, n_grid: int = 501):
        x_grid = self._xgrid_from_zero(n_peak, n_grid)
        f = self._pd.pair_distribution_function(x_grid, n_peak)
        ax.plot(x_grid, f, label=r'$G(r) = \frac{R(r)}{r} - 2{\pi}\rho_0$')
        ax.set_xlabel("r")
        ax.set_ylabel("G")
        
    def radial_distribution_function(self, ax: Axes, n_peak: int, n_grid: int = 501):
        x_grid = self._xgrid_from_zero(n_peak, n_grid)
        f = self._pd.radial_distribution_function(x_grid, n_peak)
        ax.plot(x_grid, f, label=r'$R(r) = \frac{1}{N} \sum_{u} \sum_{v} \frac{f_u f_v}{\bar{f}^2} u(r - r_{uv})$')
        ax.set_xlabel("r")
        ax.set_ylabel("R")
        
    def linear_baseline(self, ax: Axes, xmax: float):
        x = np.linspace(0, xmax, 2)
        ax.plot(x, self._pd.linear_baseline(x), '--', label=r'R = $2{\pi}{\rho}_{0}r$')
        
    def constant_baseline(self, ax: Axes, xmax: float):
        x = np.linspace(0, xmax, 2)
        ax.plot(x, np.zeros_like(x), '--', label=r'$G = 0$')
        
    def fake_fit(self, ax0: Axes, ax1: Axes, noise: float, n_peak: int, n_grid: int = 501):
        x_grid = self._xgrid_from_zero(n_peak, n_grid)
        f = self._pd.pair_distribution_function(x_grid, n_peak)
        f_noise = noise * np.random.randn(*x_grid.shape)
        ax0.plot(x_grid, f + f_noise, label=r"Observed")
        ax0.plot(x_grid, f, label=r"Calculated")
        ax0.set_xlabel("r")
        ax0.set_ylabel("G")
        ax1.hlines(0., xmin=x_grid[0], xmax=x_grid[-1], ls='--', color='gray')
        ax1.plot(x_grid, f_noise, label=r"Residuals")
        ax1.set_xlabel("r")
        ax1.set_ylabel(r"${\Delta}$G")
```

## Counting pairs

Every atom is equivalent so we just need to choose one atom as the origin and count the neigbours.

Every atom has the same scattering factor so the weights are 1.

$$
\large R(r)
= \frac{1}{N} \sum_{u} \sum_{v} \frac{f_u f_v}{\bar{f}^2} \delta(r - r_{uv})
= \sum_{v} \delta(r - r_{0v})
$$

```python
fig = plt.figure(figsize=(6 + 8, 6))
grids = GridSpec(1, 2, width_ratios=[6, 8])
ax0 = fig.add_subplot(grids[0])
ax1 = fig.add_subplot(grids[1])
sc.visualize(ax0, s=100)
sc.circles(ax0, 6, linewidth=2)
sc.radial_distribution_delta(ax1, 6)
plt.show()
```

## Uncertainty in positions

The atoms randomly deviate from the lattice sites and the deviation follows normal distribution. The peaks change from delta functions to Gaussian functions.

$$\large R(r) = \sum_{v} \frac{1}{\sigma\sqrt{2\pi}} \exp(-\frac{(r - r_{0v})^2}{2\sigma^2})$$

```python
fig = plt.figure(figsize=(8 + 8, 6))
grids = GridSpec(1, 2)
ax0 = fig.add_subplot(grids[0])
ax1 = fig.add_subplot(grids[1])
sc.radial_distribution_delta(ax0, 6)
sc.radial_distribution_gaussian(ax1, 6)
plt.show()
```

## Modeling of G(r): cost function

Optimization Problem:

$$\underset{p_0, p_1, ... , p_n}{\mathrm{argmin}} \frac{1}{2} \chi^2(p_0, p_1, ... , p_n)$$

Calculate cost function from residuals:

$$\chi^2(p_0, p_1, ... , p_n) = \int_0^{\infty} \left[ \frac{G_{obs}(r) - G_{cal}(r, p_0, p_1, ... , p_n)}{\Delta G_{obs}(r)} \right] ^2 dr$$

The $G_{cal}(r)$ is determined by structure models:

$$G_{cal}(r, p_0, p_1, ... , p_n) = F(Structure(p_0, p_1, ... , p_n))$$

## Modeling of G(r): visualize $\chi^2$

Usually, we don't know $\Delta G_{obs}(r)$ so we assume $\Delta G_{obs}(r) = 1$. Usually, the residuals oscillate around zero. The $\chi^2$ is manifested by the amplitude of oscillation.

```python
sc = SquareCrystal(3, 1., 0.05)
fig = plt.figure(figsize=(8, 6))
grids = GridSpec(2, 1, height_ratios=[8, 2], hspace=0)
ax0 = fig.add_subplot(grids[0])
ax0.set_xlabel(None)
ax0.set_xticklabels([])
ax1 = fig.add_subplot(grids[1])
sc.fake_fit(ax0, ax1, 0.3, 6)
ax0.legend()
plt.show()
del sc, fig, grids, ax0, ax1
```
